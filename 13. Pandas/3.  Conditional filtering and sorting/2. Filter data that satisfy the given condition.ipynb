### Combining two filters

```python
filter1 = df['account length'] > 100
filter2 = df['total night calls'] < 90
```

```python
df.loc[ (filter1 | filter2), : ]
```

```python
df.loc[ (filter1 & filter2), : ]
```

### where

Instead of droppping the non-eligible rows, where method replaces the value using NAN

```python
df.where(filter1 & filter2).head(5)
```

#### You can either drop these rows or replace the NANs with some other values

```python
df.where(filter1 & filter2).fillna(0)

# Filling missing value with 0
```

```python
df.where(filter1 & filter2).dropna()

#dropping the row containing missing values
```
