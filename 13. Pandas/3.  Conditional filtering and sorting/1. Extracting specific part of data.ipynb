##### Pandas offers 5 primary methods to select items:

 1] Dot notation : Select a single column

 2] loc  : location : select data based on label (row name , column name)

 3] iloc : integer location - index based

 4] at  : select one item only based on row and column label

 5] iat   : select one item only based on row and column number

### Dot Notation

Select one column only as reference

Syntax - dataframe.column_name

```python
df.state
```

The object returned on selecting just one column is a pandas series

```python
type(df.state)
```

## .loc


loc - label based - selects data based on labels or a boolean array

loc - used for selecting data from a dataframe

.loc takes 2 arguments inside the square brackets.

1] row name - label

2] column name - header

##### Syntax - dataframe.loc['row_name', 'column_name']

When slicing, It doesn't include end value

#### Selecting single row by label

```python
df.head()
```

```python
df.loc[1]
```

#### Selecting multiple rows by label

```python
df.loc[[1,2]]
```

#### Selecting rows and specific columns by labels

```python
df.loc[1, ['area code', 'state']]
```

#### Slicing rows and columns by labels:



```python
df.loc[1:5, ['state', 'area code']]
```

## .iloc

Integer position-based: select data based on index.

Syntax: df.iloc[rows, columns]

Does not include end: When slicing, it excludes the end value.


#### Selecting a single row by index

```python
df.head()
```

```python
df.iloc[0]
```

#### Selecting multiple rows by index:

```python
df.iloc[[0,1]]   # First row , Second row
```

#### Selecting rows and specific columns by index:



```python
df.iloc[0, [0,1,2]]
```

#### Slicing rows and columns by index

```python
df.iloc[0:3 , 0:4]
```

## Comparison

loc is generally used when you want to work with labels, which are more meaningful and easier to understand.

iloc is generally used when you want to work with position   


### at

Label based scalar accessor

It is used to access a single value using row label & column label

df.at[row_label, column_label]

scalar  - single value / single element

```python
df.head()
```

```python
df.iat[1, 1]
```

```python
df.iat[2,1]
```

### .iat

integer location based scalar accessor

It is used to access a single value using row indices & column indices

df.at[row_indices, column_indices]

```python
df.head()
```

```python
df.iat[0,2]
```
