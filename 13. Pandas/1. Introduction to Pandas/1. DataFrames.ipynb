```python
import numpy as np
import pandas as pd
```

```python
arr = np.random.randint(0,100,(5,3))
arr
```

### Creating dataFrame from numpy array

```python
# we use pd.DataFrame(numpyarray)
df = pd.DataFrame(arr)
df
```

```python
df= pd.DataFrame(arr, index=['Mon','Tues','Wed','Thurs','Fri'], columns=['Jan','Feb','Mar'])
df
```

```python
# Custom rows and Columns
rowname = ['Mon','Tues','Wed','Thurs','Fri']
columnsname=['Jan','Feb','Mar']
```

```python
df= pd.DataFrame(arr, index=rowname, columns=columnsname)
df
```

### Creating DataFrame from dictionary

```python
mydict = {
    'Jan' : [1,2,3,4,5],
    'Feb' : [10,20,30,40,50],
    'Mar' : [60,70,80,90,100]
}

mydict
```

```python
df = pd.DataFrame(mydict)
df
```

```python
#To get columns in specific order, explicitly mention it
```

```python
df = pd.DataFrame(mydict, index=['Mon','Tues','Wed','Thurs','Fri'], columns = ['Jan','Feb','Mar'] )
df
```
