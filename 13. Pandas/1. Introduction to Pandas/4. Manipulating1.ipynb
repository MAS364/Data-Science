```python
df.head()
```

### drop()

- To remove the row , we use drop function
- Pass the row that you want to delete
- Takes a label, not position
- Drop function doesn't do it inplace, It make a copy of original dataframe without including the row you want to delete
- If you want change to happen in original dataframe itself, pass argument inplace = True, by default inplace is false

```python
df.drop(0)
df.head()

# Drop function doesn't do it inplace, It make a copy of original dataframe without including the row you want to delete
```

```python
a = df.drop(0)
a.head()
```

```python
df.drop(0, inplace = True)
df.head()

# Here, we are not passing index 0, we are not saying remove the 0th row
# Here, we are saying remove the row which has label zero
# Label stick to the row
```

```python
df.drop(3, inplace = True)
df.head()
```
