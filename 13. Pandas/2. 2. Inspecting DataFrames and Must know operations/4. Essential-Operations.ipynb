### To get relative frequencies

```python
# apple: Appears 4 times out of 10, so the relative frequency is  4 / 10 = 0.4
# banana: Appears 4 times out of 10, so the relative frequency is 4 / 10 = 0.4
# orange: Appears 2 times out of 10, so the relative frequency is 2 / 10 = 0.4
```

```python
df['state'].value_counts(normalize = True)
```

```python
df['state'].value_counts(normalize = True) * 100
```

### Get first 'n' rows order by a given column

```python
# Syntax - DataFrame.nlargest(n, columns, keep='first')

# The df.nlargest() method in pandas is used to retrieve the rows of a DataFrame 
# with the largest values in a specified column. Here’s how you can use it:

# n: Number of rows to return.
# columns: Column label or list of column labels to order by.
# keep: Determines which duplicates (if any) to keep. The default is 'first', 
# which keeps the first occurrence. Options are 'first', 'last', and False.
```

```python
df.nlargest(15, 'account length')
```
