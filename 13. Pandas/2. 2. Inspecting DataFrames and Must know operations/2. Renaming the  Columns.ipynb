### Rename all columns in one shot :  Replace all space character with "_"

#### Use lambda function

```python
df.rename(lambda x: x.replace(" ","_"), axis='columns').head()

#x is a parameter that takes column name of all the columns present in the dataframe
```

### Change all column name to title case : " account length " : Account Length 

```python
df.rename(lambda x: x.title(), axis='columns').head()
```
